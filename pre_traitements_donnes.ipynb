```python
# Function to calculate missing values by column# Funct 
def missing_values_table(df):
        # Total missing values
        mis_val = df.isnull().sum()
        
        # Percentage of missing values
        mis_val_percent = 100 * df.isnull().sum() / len(df)
        
        # Make a table with the results
        mis_val_table = pd.concat([mis_val, mis_val_percent], axis=1)
        
        # Rename the columns
        mis_val_table_ren_columns = mis_val_table.rename(
        columns = {0 : 'Missing Values', 1 : '% of Total Values'})
        
        # Sort the table by percentage of missing descending
        mis_val_table_ren_columns = mis_val_table_ren_columns[
            mis_val_table_ren_columns.iloc[:,1] != 0].sort_values(
        '% of Total Values', ascending=False).round(1)
        
        # Print some summary information
        print ("Your selected dataframe has " + str(df.shape[1]) + " columns.\n"      
            "There are " + str(mis_val_table_ren_columns.shape[0]) +
              " columns that have missing values.")
        
        # Return the dataframe with missing information
        return mis_val_table_ren_columns
```

```python
def traitement_app_train():
 
    #Traitement de application_train
    data = pd.read_csv('D:/openclassrooms/PROJET 7/DATA/application_train.csv')
    
    #données selectionnées
    col=['SK_ID_CURR','TARGET','NAME_CONTRACT_TYPE','CODE_GENDER','FLAG_OWN_CAR',
         'FLAG_OWN_REALTY','CNT_CHILDREN','AMT_INCOME_TOTAL','AMT_CREDIT','AMT_GOODS_PRICE',
         'AMT_ANNUITY','NAME_INCOME_TYPE','NAME_EDUCATION_TYPE','NAME_FAMILY_STATUS','NAME_HOUSING_TYPE',
         'DAYS_BIRTH','DAYS_EMPLOYED','CNT_FAM_MEMBERS','REGION_POPULATION_RELATIVE','REGION_RATING_CLIENT','REGION_RATING_CLIENT_W_CITY',
         'EXT_SOURCE_1','EXT_SOURCE_2','EXT_SOURCE_3',
         'OBS_30_CNT_SOCIAL_CIRCLE','DEF_30_CNT_SOCIAL_CIRCLE','OBS_60_CNT_SOCIAL_CIRCLE','DEF_60_CNT_SOCIAL_CIRCLE',
          'DAYS_LAST_PHONE_CHANGE','FLAG_DOCUMENT_2','FLAG_DOCUMENT_3', 'FLAG_DOCUMENT_4', 'FLAG_DOCUMENT_5',
       'FLAG_DOCUMENT_6', 'FLAG_DOCUMENT_7', 'FLAG_DOCUMENT_8',
       'FLAG_DOCUMENT_9', 'FLAG_DOCUMENT_10', 'FLAG_DOCUMENT_11',
       'FLAG_DOCUMENT_12', 'FLAG_DOCUMENT_13', 'FLAG_DOCUMENT_14',
       'FLAG_DOCUMENT_15', 'FLAG_DOCUMENT_16', 'FLAG_DOCUMENT_17',
       'FLAG_DOCUMENT_18', 'FLAG_DOCUMENT_19', 'FLAG_DOCUMENT_20']
   
    col_flag = ['FLAG_DOCUMENT_2',
       'FLAG_DOCUMENT_3', 'FLAG_DOCUMENT_4', 'FLAG_DOCUMENT_5',
       'FLAG_DOCUMENT_6', 'FLAG_DOCUMENT_7', 'FLAG_DOCUMENT_8',
       'FLAG_DOCUMENT_9', 'FLAG_DOCUMENT_10', 'FLAG_DOCUMENT_11',
       'FLAG_DOCUMENT_12', 'FLAG_DOCUMENT_13', 'FLAG_DOCUMENT_14',
       'FLAG_DOCUMENT_15', 'FLAG_DOCUMENT_16', 'FLAG_DOCUMENT_17',
       'FLAG_DOCUMENT_18', 'FLAG_DOCUMENT_19', 'FLAG_DOCUMENT_20']

    data = data[col]
    
    #creation d'une variable sommant les valeurs EXT_SOURCE
    data['EXT_SOURCE_SUM'] = data['EXT_SOURCE_1']+data['EXT_SOURCE_2']+data['EXT_SOURCE_2']
    #creation d'une variable moyennant les valeurs EXT_SOURCE
    data['EXT_SOURCE_MEAN'] = (data['EXT_SOURCE_1']+data['EXT_SOURCE_2']+data['EXT_SOURCE_2'])/3
    #supression des 4 clients ayant un genre non determiné
    data.drop( data[ data['CODE_GENDER'] == "XNA" ].index , inplace=True)
    #data.drop( data[ data['NAME_CONTRACT_TYPE'] == "Revolving loans" ].index , inplace=True)
    #creation d'une variable longueur du credit en mois
    data['LENGTH_CREDIT']=data['AMT_CREDIT']/data['AMT_ANNUITY']
    #suppression valeur aberrante AMT_INCOME_TOTAL
    data.drop( data[ data['AMT_INCOME_TOTAL'] == 117000000.0 ].index , inplace=True)
    #suppression des 12 valeurs manquantes de AMT_ANNUITY
    data.drop( data[ data['AMT_ANNUITY'].isnull() ].index , inplace=True)
    #Creation de la variable INCOME_PER_PERSON
    data['INCOME_PER_PERSON'] = data['AMT_INCOME_TOTAL'] / data['CNT_FAM_MEMBERS']
    data['REMAIN_FOR_LIFE'] = data['AMT_INCOME_TOTAL'] - data['AMT_ANNUITY']
    data['REMAIN_FOR_LIFE_PER_PERSON'] = data['REMAIN_FOR_LIFE'] / data['CNT_FAM_MEMBERS']
    #Creation de la variable CREDIT_INCOME_RATIO Montant du credit divisé par le revenu mensuel
    data['CREDIT_INCOME_RATIO'] = data['AMT_CREDIT'] / data['AMT_INCOME_TOTAL']
    #creation ANNUITY_INCOME_RATIO part de la mensualité par rapport revenu
    data['ANNUITY_INCOME_RATIO'] = data['AMT_ANNUITY']/ data['AMT_INCOME_TOTAL']
    #DAYS_EMPLOYED est égal à 365243 quand les demandeurs sont "Pensioner" ou "Unemployed", la valeur devient 0
    data["DAYS_EMPLOYED"][data["DAYS_EMPLOYED"] == 365243 ] = 0
    #creation de YEARS_EMPLOYED qui compte les années d'ancienneté variable float car des clients ont moins d'un an
    data['YEARS_EMPLOYED'] = data['DAYS_EMPLOYED']/365*(-1)
    #creation de YEARS_BIRTH pour plus de visibilité 
    data['YEARS_BIRTH'] = data['DAYS_BIRTH']/365*(-1)
    #creation de EMPLOYED_RATIO evalue la carriere professionnelle dans l'entreprise en ancienneté 
    data['EMPLOYED_RATIO'] = data['DAYS_EMPLOYED'] / data['DAYS_BIRTH']
    # suppression de la valeur aberrante CNT_SOCIAL_CIRCLE
    data.drop( data[data['OBS_30_CNT_SOCIAL_CIRCLE']>50].index , inplace=True)
    #creation de DEF_OBS_30 et DEF_OBS_60 ratio observation faisant defaut par observation totales
    data['DEF_OBS_30'] = data['DEF_30_CNT_SOCIAL_CIRCLE'] / data['OBS_30_CNT_SOCIAL_CIRCLE']
    data['DEF_OBS_30'] = data['DEF_OBS_30'].fillna(0)
    data['DEF_OBS_60'] = data['DEF_60_CNT_SOCIAL_CIRCLE'] / data['OBS_60_CNT_SOCIAL_CIRCLE']
    data['DEF_OBS_60'] = data['DEF_OBS_30'].fillna(0)
    #creation de RATE_DOC taux de documents fourni par rapport à20, nombre de documents 
    data['RATE_DOC'] = data[col_flag].sum(axis=1)/20
    
    data = data.drop(col_flag, axis=1)
    data = data.drop(['DAYS_EMPLOYED','DAYS_BIRTH'], axis=1)
    data = data[data.NAME_CONTRACT_TYPE == "Cash loans"]           
    data = data.drop(['NAME_CONTRACT_TYPE'], axis=1)
    
    data['OBS_30_CNT_SOCIAL_CIRCLE'] = data['OBS_30_CNT_SOCIAL_CIRCLE'].fillna(0)
    data['OBS_60_CNT_SOCIAL_CIRCLE'] = data['OBS_60_CNT_SOCIAL_CIRCLE'].fillna(0)
    data['DEF_30_CNT_SOCIAL_CIRCLE'] = data['DEF_30_CNT_SOCIAL_CIRCLE'].fillna(0)
    data['DEF_60_CNT_SOCIAL_CIRCLE'] = data['DEF_60_CNT_SOCIAL_CIRCLE'].fillna(0)    
    data['DAYS_LAST_PHONE_CHANGE'] = data['DAYS_LAST_PHONE_CHANGE'].fillna(0)  
    
    return(data)
```

```python
def traitement_bureau(data):

    bureau = pd.read_csv('D:/openclassrooms/PROJET 7/DATA/bureau.csv')
    #*****Aggregation de bureau et bureau_balance *********
    
    # Creation des variables "BUR_TOT_CREDIT','BUR_TOT_CREDIT_DEBT', 'BUR_TOT_CREDIT_OVERDUE','BUR_TOT_CREDIT_PROLONG',
    #'BUR_TOT_ANNUITY', 'BUR_MIN_DAYS_CREDIT', 'BUR_MAX_DAYS_OVERDUE'
    
    bureau_agr1 = bureau.groupby('SK_ID_CURR', as_index = False).agg({'AMT_CREDIT_SUM': 'sum',
                                                                      'AMT_CREDIT_SUM_DEBT':'sum',
                                                                      'AMT_CREDIT_SUM_OVERDUE':"sum",
                                                                      'CNT_CREDIT_PROLONG':'sum', 
                                                                      'AMT_ANNUITY': 'sum',
                                                                      'DAYS_CREDIT' : 'min',
                                                                      'CREDIT_DAY_OVERDUE': 'max',
                                                                     })

                
    bureau_agr1.columns = ['SK_ID_CURR', 'BUR_TOT_CREDIT','BUR_TOT_CREDIT_DEBT', 'BUR_TOT_CREDIT_OVERDUE',
                         'BUR_TOT_CREDIT_PROLONG', 'BUR_TOT_ANNUITY', 'BUR_MIN_DAYS_CREDIT', 
                         'BUR_MAX_DAYS_OVERDUE']
    data = data.merge(bureau_agr1, on = 'SK_ID_CURR', how = 'left') # Join to the data dataframe
    data[bureau_agr1.columns] = data[bureau_agr1.columns].fillna(0) # Fill the missing values with 0 
    

    # Creation de BUR_CNT_ACTIVE_LOAN nombre de credits actif
    bureau_agr2 = bureau[bureau.CREDIT_ACTIVE == 'Active'].groupby('SK_ID_CURR', as_index=False)['SK_ID_BUREAU'].count()
    bureau_agr2.columns = ['SK_ID_CURR','BUR_CNT_ACTIVE_LOAN']
    data = data.merge(bureau_agr2, on = 'SK_ID_CURR', how = 'left') # Join to the data dataframe
    data['BUR_CNT_ACTIVE_LOAN'] = data['BUR_CNT_ACTIVE_LOAN'].fillna(0) # Fill the missing values with 0 
    
    # Creation de BUR_CNT_CLOSED_LOAN nombre de credits fermes ou soldés
    bureau_agr3 = bureau[(bureau.CREDIT_ACTIVE == 'Closed')|(bureau.CREDIT_ACTIVE == 'Sold')].groupby('SK_ID_CURR', as_index=False)['SK_ID_BUREAU'].count()
    bureau_agr3.columns = ['SK_ID_CURR','BUR_CNT_CLOSED_LOAN']
    data = data.merge(bureau_agr3, on = 'SK_ID_CURR', how = 'left') # Join to the data dataframe
    data['BUR_CNT_CLOSED_LOAN'] = data['BUR_CNT_CLOSED_LOAN'].fillna(0) # Fill the missing values with 0 
    
    # Creation de BUR_SUM_ANNUITY_ACTIVE et BUR_TOT_ANNUITY_ACTIVE
    bureau_agr4 = bureau[bureau.CREDIT_ACTIVE == 'Active'].groupby('SK_ID_CURR', as_index=False).agg({'AMT_ANNUITY':'sum','AMT_CREDIT_SUM': 'sum'})
    bureau_agr4.columns = ['SK_ID_CURR','BUR_TOT_ANNUITY_ACTIVE','BUR_TOT_CREDIT_ACTIVE']
    data = data.merge(bureau_agr4, on = 'SK_ID_CURR', how = 'left') # Join to the data dataframe
    data[bureau_agr4.columns] = data[bureau_agr4.columns].fillna(0) # Fill the missing values with 0    
    
    # Creation de BUR_CNT_REQUEST_1Y
    bureau_agg5 = bureau[bureau.DAYS_CREDIT>= -365].groupby('SK_ID_CURR', as_index=False)['SK_ID_BUREAU'].count()
    bureau_agg5.columns = ['SK_ID_CURR','BUR_CNT_REQUEST_1Y']
    data = data.merge(bureau_agg5, on = 'SK_ID_CURR', how = 'left') # Join to the data dataframe
    data['BUR_CNT_REQUEST_1Y'] = data['BUR_CNT_REQUEST_1Y'].fillna(0) # Fill the missing values with 0   
    
    bureau
    
    bureau_balance =  pd.read_csv('D:/openclassrooms/PROJET 7/DATA/bureau_balance.csv')
    
    #***********Aggregation des données de la table bureau_balance*****
    
    #Creation de la variable 'BUR_BAL_STATUS_BAD_1Y' Nombre de mois avec mauvais DPD >60 J dans la derniere année
    bureau_balance_agg= pd.get_dummies(bureau_balance.STATUS,prefix='STATUS')
    bureau_balance_agg = pd.concat([bureau_balance, bureau_balance_agg], axis=1)
    
    #somme des status 
    bureau_balance_agg['DPD_BAD'] = bureau_balance_agg.STATUS_3 + bureau_balance_agg.STATUS_4 + bureau_balance_agg.STATUS_5
    
    bureau_balance_agg1 = bureau_balance_agg[bureau_balance_agg.MONTHS_BALANCE >= -12] #selection derniere année
    
    bureau_balance_agg1 = bureau_balance_agg1[['SK_ID_BUREAU','DPD_BAD']].groupby('SK_ID_BUREAU',as_index=False).agg({'DPD_BAD':'sum'})
    bureau_balance_agg1 = bureau[['SK_ID_BUREAU', 'SK_ID_CURR']].merge(bureau_balance_agg1, on = 'SK_ID_BUREAU', how = 'left')
    bureau_balance_agg1 = bureau_balance_agg1.groupby('SK_ID_CURR',as_index=False).agg({'DPD_BAD':'sum'})
    bureau_balance_agg1.columns = ['SK_ID_CURR','BUR_BAL_DPD_BAD_1Y']
    data = data.merge(bureau_balance_agg1, on = 'SK_ID_CURR', how = 'left')
    data['BUR_BAL_DPD_BAD_1Y'] = data.BUR_BAL_DPD_BAD_1Y.fillna(0)
    
    bureau_balance_agg = bureau_balance_agg[bureau_balance_agg.MONTHS_BALANCE >= -36] #selection 3 dernieres années
    bureau_balance_agg = bureau_balance_agg[['SK_ID_BUREAU','DPD_BAD']].groupby('SK_ID_BUREAU',as_index=False).agg({'DPD_BAD':'sum'})
    bureau_balance_agg = bureau[['SK_ID_BUREAU', 'SK_ID_CURR']].merge(bureau_balance_agg, on = 'SK_ID_BUREAU', how = 'left')
    bureau_balance_agg = bureau_balance_agg.groupby('SK_ID_CURR',as_index=False).agg({'DPD_BAD':'sum'})
    bureau_balance_agg.columns = ['SK_ID_CURR','BUR_BAL_DPD_BAD_3Y']
    data = data.merge(bureau_balance_agg, on = 'SK_ID_CURR', how = 'left')
    data['BUR_BAL_DPD_BAD_3Y'] = data.BUR_BAL_DPD_BAD_3Y.fillna(0)
    
    del bureau
    del bureau_balance
    
    return data
```

```python
def traitement_previous_application (data):


    previous_application = pd.read_csv('D:/openclassrooms/PROJET 7/DATA/previous_application.csv')

  #*************Aggregation de la table previous_application¶********
    
    #Creation de la variable CNT_INSURED_ON nb assurances souscrites par le clients lors de ses crédits précédents chez Home Credit
    previous_application_agg = previous_application.groupby('SK_ID_CURR', as_index=False)['NFLAG_INSURED_ON_APPROVAL'].sum().rename(columns={'NFLAG_INSURED_ON_APPROVAL': 'CNT_INSURED_ON'})
    data = data.merge(previous_application_agg, on = 'SK_ID_CURR', how = 'left')
    data['CNT_INSURED_ON'] = data['CNT_INSURED_ON'].fillna(0)
    
    #Creation des variables CNT_LOANS_REFUSED, CNT_LOANS_APPROVED, CNT_LOANS_CANCELED
    #nombre de crédits precedent approuvés et refusés chez home credit
    previous_application_agg = previous_application[previous_application.NAME_CONTRACT_STATUS == "Approved"].groupby('SK_ID_CURR', as_index=False)['SK_ID_PREV'].count().rename(columns={'SK_ID_PREV': 'CNT_LOANS_APPROVED'})
    data = data.merge(previous_application_agg, on = 'SK_ID_CURR', how = 'left')
    data['CNT_LOANS_APPROVED'] = data['CNT_LOANS_APPROVED'].fillna(0)
    previous_application_agg = previous_application[previous_application.NAME_CONTRACT_STATUS == "Canceled"].groupby('SK_ID_CURR', as_index=False)['SK_ID_PREV'].count().rename(columns={'SK_ID_PREV': 'CNT_LOANS_CANCELED'})
    data = data.merge(previous_application_agg, on = 'SK_ID_CURR', how = 'left')
    data['CNT_LOANS_CANCELED'] = data['CNT_LOANS_CANCELED'].fillna(0)    
    previous_application_agg = previous_application[previous_application.NAME_CONTRACT_STATUS == "Refused"].groupby('SK_ID_CURR', as_index=False)['SK_ID_PREV'].count().rename(columns={'SK_ID_PREV': 'CNT_LOANS_REFUSED'})
    data = data.merge(previous_application_agg, on = 'SK_ID_CURR', how = 'left')
    data['CNT_LOANS_REFUSED'] = data['CNT_LOANS_REFUSED'].fillna(0)
    
    #Creation de la variable CNT_ACTIVE_LOAN nombre de credit encore actifs chez Home Credit
    # et des variables CNT_ACTIVE_REV_LOAN, CNT_ACTIVE_CONS_LOAN, CNT_ACTIVE_CASH_LOAN suivant le type de credit
    # si DAYS_TERMINATION >0 et DAYS_LAST_DUE >0 le credit est encore actif
    previous_application['END_LOAN'] = previous_application[['DAYS_LAST_DUE','DAYS_TERMINATION']].min(axis=1)
    previous_application_agg = previous_application[previous_application.END_LOAN >0].groupby('SK_ID_CURR', as_index=False)['SK_ID_PREV'].count().rename(columns={'SK_ID_PREV': 'CNT_ACTIVE_LOAN'})
    data = data.merge(previous_application_agg, on = 'SK_ID_CURR', how = 'left')
    data['CNT_ACTIVE_LOAN'] = data['CNT_ACTIVE_LOAN'].fillna(0)
    
    #test = previous_application[(previous_application.END_LOAN >0) & (previous_application.NAME_CONTRACT_TYPE = 'Revolving loans')]
    previous_application_agg = previous_application[(previous_application.END_LOAN >0)&(previous_application.NAME_CONTRACT_TYPE == 'Revolving loans')].groupby('SK_ID_CURR', as_index=False)['SK_ID_PREV'].count().rename(columns={'SK_ID_PREV': 'CNT_ACTIVE_REV_LOAN'})
    
    data = data.merge(previous_application_agg, on = 'SK_ID_CURR', how = 'left')
    data['CNT_ACTIVE_REV_LOAN'] = data['CNT_ACTIVE_REV_LOAN'].fillna(0)    
    previous_application_agg = previous_application[(previous_application.END_LOAN >0)&(previous_application.NAME_CONTRACT_TYPE == 'Consumer loans')].groupby('SK_ID_CURR', as_index=False)['SK_ID_PREV'].count().rename(columns={'SK_ID_PREV': 'CNT_ACTIVE_CONS_LOAN'})
    data = data.merge(previous_application_agg, on = 'SK_ID_CURR', how = 'left')
    data['CNT_ACTIVE_CONS_LOAN'] = data['CNT_ACTIVE_CONS_LOAN'].fillna(0)    
    previous_application_agg = previous_application[(previous_application.END_LOAN >0)&(previous_application.NAME_CONTRACT_TYPE == 'Cash loans')].groupby('SK_ID_CURR', as_index=False)['SK_ID_PREV'].count().rename(columns={'SK_ID_PREV': 'CNT_ACTIVE_CASH_LOAN'})
    data = data.merge(previous_application_agg, on = 'SK_ID_CURR', how = 'left')
    data['CNT_ACTIVE_CASH_LOAN'] = data['CNT_ACTIVE_CASH_LOAN'].fillna(0) 
    
    #Creation de TOT_ACTIVE_ANNUITY montant des mensualités encore actives chez Home Credit, tout crédit confondu
    #Creation de 
    previous_application_agg = previous_application[previous_application.END_LOAN > 0].groupby('SK_ID_CURR', as_index = False)['AMT_ANNUITY'].sum().rename(columns = {'AMT_ANNUITY': 'TOT_ACTIVE_ANNUITY'})
    data = data.merge(previous_application_agg, on = 'SK_ID_CURR', how = 'left')
    data['TOT_ACTIVE_ANNUITY'] = data['TOT_ACTIVE_ANNUITY'].fillna(0) 
    
    #Creation de MEAN_CNT_PAYMENT_FUTURE moyenne de jours restants par credit actifs hors revolvinf
    #Calcul des jours restant de fin de credit hors revolving loan (qui ont CNT_PAYMENT = 0)
    #PREV_END_LOAN = longueur totale du credit prevu en jours - le nombre de jour depuis la decision de credit
    previous_application['PREV_END_LOAN'] = previous_application.CNT_PAYMENT*30 + previous_application.DAYS_DECISION
    #Moyenne des valeurs obtenues pour les credits actifs (END_LOAN >0) et les credits hors revolving
    previous_application_agg = previous_application[(previous_application.END_LOAN >0) & (previous_application.CNT_PAYMENT >0) ].groupby('SK_ID_CURR', as_index=False)['PREV_END_LOAN'].mean().rename(columns={'PREV_END_LOAN': 'MEAN_CNT_PAYMENT_FUTURE'})
    data = data.merge(previous_application_agg, on = 'SK_ID_CURR', how = 'left')
    data['MEAN_CNT_PAYMENT_FUTURE'] = data['MEAN_CNT_PAYMENT_FUTURE'].fillna(0)
    
    del previous_application
    
    return data
```

```python
def traitement_installments_payments(data):

    installments_payments = pd.read_csv('D:/openclassrooms/PROJET 7/DATA/installments_payments.csv')

    #******Aggregation de installments_payments.csv********************************************************************
    #Creation des variables de retard de paiement et de defaut de paiement des mensualité lors de la derniere année
    #et lors des 3 dernieres années
    
    #Retard de paiement par mensualité nb de jour de retard (valeur negative si il paye en avance)
    installments_payments['DPD'] = installments_payments.DAYS_ENTRY_PAYMENT -installments_payments.DAYS_INSTALMENT
    #transformation en variables binaires pour les retards >60jours
    installments_payments['DPD_60'] = (installments_payments['DPD']>60).astype(int)
    #calcul sur N-1
    installments_payments_agg= installments_payments[installments_payments.DAYS_INSTALMENT >= -365]
    installments_payments_agg = installments_payments_agg.groupby('SK_ID_CURR', as_index=False)['DPD_60'].sum().rename(columns={'DPD_60': 'CNT_DPD_60_1Y'})
    data = data.merge(installments_payments_agg, on = 'SK_ID_CURR', how = 'left')    
    data['CNT_DPD_60_1Y'] = data['CNT_DPD_60_1Y'].fillna(0)
    #calcul sur N-3
    installments_payments_agg= installments_payments[installments_payments.DAYS_INSTALMENT >= -365*3]
    installments_payments_agg = installments_payments_agg.groupby('SK_ID_CURR', as_index=False)['DPD_60'].sum().rename(columns={'DPD_60': 'CNT_DPD_60_3Y'})
    data = data.merge(installments_payments_agg, on = 'SK_ID_CURR', how = 'left')    
    data['CNT_DPD_60_3Y'] = data['CNT_DPD_60_3Y'].fillna(0)
    
    #Creation des variables CNT_LESS_PAYMENT qui compte les mensualités payées avec au moins -30% du montant prévu
    #sur N-1 et N-3
    installments_payments['LESS_PAYMENT']=(installments_payments.AMT_PAYMENT<(installments_payments.AMT_INSTALMENT*0.70)).astype(int)
    #sur N-1    
    installments_payments_agg= installments_payments[installments_payments.DAYS_INSTALMENT >= -365]
    installments_payments_agg = installments_payments_agg.groupby('SK_ID_CURR', as_index=False)['LESS_PAYMENT'].sum().rename(columns={'LESS_PAYMENT': 'CNT_LESS_PAYMENT_1Y'})
    data = data.merge(installments_payments_agg, on = 'SK_ID_CURR', how = 'left')    
    data['CNT_LESS_PAYMENT_1Y'] = data['CNT_LESS_PAYMENT_1Y'].fillna(0)    
    #sur N-3
    installments_payments_agg= installments_payments[installments_payments.DAYS_INSTALMENT >= -365*3]
    installments_payments_agg = installments_payments_agg.groupby('SK_ID_CURR', as_index=False)['LESS_PAYMENT'].sum().rename(columns={'LESS_PAYMENT': 'CNT_LESS_PAYMENT_3Y'})
    data = data.merge(installments_payments_agg, on = 'SK_ID_CURR', how = 'left')    
    data['CNT_LESS_PAYMENT_3Y'] = data['CNT_LESS_PAYMENT_3Y'].fillna(0)     
    
    del installments_payments
    
    return data
```

```python
def traitement_POS_CASH_balance(data):

    POS_CASH_balance = pd.read_csv('D:/openclassrooms/PROJET 7/DATA/POS_CASH_balance.csv')
    
     #****Aggregation de POS_CASH_balance ***********************************************************************
    #Creation des variables de retard de paiement + 60j et de defaut de paiement des mensualité lors de la derniere année
    #et lors des 3 dernieres années
    POS_CASH_balance['DPD_60'] = (POS_CASH_balance['SK_DPD']>60).astype(int)
   
    #calcul sur N-1
    POS_CASH_balance_agg= POS_CASH_balance[POS_CASH_balance.MONTHS_BALANCE >= -12]
    POS_CASH_balance_agg = POS_CASH_balance_agg.groupby('SK_ID_CURR', as_index=False)['DPD_60'].sum().rename(columns={'DPD_60': 'POS_CNT_DPD_60_1Y'})
    data = data.merge(POS_CASH_balance_agg, on = 'SK_ID_CURR', how = 'left')    
    data['POS_CNT_DPD_60_1Y'] = data['POS_CNT_DPD_60_1Y'].fillna(0)
    #calcul sur N-3
    POS_CASH_balance_agg= POS_CASH_balance[POS_CASH_balance.MONTHS_BALANCE >= -12]
    POS_CASH_balance_agg = POS_CASH_balance_agg.groupby('SK_ID_CURR', as_index=False)['DPD_60'].sum().rename(columns={'DPD_60': 'POS_CNT_DPD_60_3Y'})
    data = data.merge(POS_CASH_balance_agg, on = 'SK_ID_CURR', how = 'left')    
    data['POS_CNT_DPD_60_3Y'] = data['POS_CNT_DPD_60_3Y'].fillna(0)
    
    #Creation de la variable POS_MEAN_CNT_INSTALMENT_FUTURE moyenne de la longueur des credits POS CASH actifs en nb de mois
    POS_CASH_balance_agg= POS_CASH_balance[(POS_CASH_balance.MONTHS_BALANCE == -1) &
                                               (POS_CASH_balance.CNT_INSTALMENT_FUTURE > 0)]
    POS_CASH_balance_agg = POS_CASH_balance_agg.groupby('SK_ID_CURR', as_index=False)['CNT_INSTALMENT_FUTURE'].mean().rename(columns={'CNT_INSTALMENT_FUTURE': 'POS_MEAN_CNT_INSTALMENT_FUTURE'})
    data = data.merge(POS_CASH_balance_agg, on = 'SK_ID_CURR', how = 'left')    
    data['POS_MEAN_CNT_INSTALMENT_FUTURE'] = data['POS_MEAN_CNT_INSTALMENT_FUTURE'].fillna(0)  
    
    del POS_CASH_balance
    
    return (data)
```

```python
def traitement_credit_card_balance(data):
    
    credit_card_balance = pd.read_csv('D:/openclassrooms/PROJET 7/DATA/credit_card_balance.csv')
    #****Aggregation de credit_card_balance
    # nombre de DPD > 60 des credits sur les 12 derniers mois
    # SK_DPD >=60
    credit_card_balance['DPD_60'] = (credit_card_balance['SK_DPD']>=60).astype(int)
    credit_card_balance_agg= credit_card_balance[credit_card_balance.MONTHS_BALANCE >= -12]
    credit_card_balance_agg = credit_card_balance_agg.groupby('SK_ID_CURR', as_index=False)['DPD_60'].sum().rename(columns={'DPD_60': 'CARD_CNT_DPD_60_1Y'})
    data = data.merge(credit_card_balance_agg, on = 'SK_ID_CURR', how = 'left')    
    data['CARD_CNT_DPD_60_1Y'] = data['CARD_CNT_DPD_60_1Y'].fillna(0)
    
    del credit_card_balance
    
    return data
    
```

```python
def agg_features(data):

    data['GLOBAL_CNT_DPD_60_1Y'] = data['CARD_CNT_DPD_60_1Y'] + data['POS_CNT_DPD_60_1Y'] + data['CNT_DPD_60_1Y'] + data['BUR_BAL_DPD_BAD_1Y']
    data['GLOBAL_CNT_DPD_60_3Y'] = data['CARD_CNT_DPD_60_1Y'] + data['POS_CNT_DPD_60_1Y'] + data['CNT_DPD_60_1Y'] + data['BUR_BAL_DPD_BAD_1Y']

    data['GLOBAL_SUM_ANNUITY'] = data['AMT_ANNUITY'] + data['BUR_TOT_ANNUITY_ACTIVE'] + data['TOT_ACTIVE_ANNUITY']
    data['GLOBAL_CREDIT_INCOME_RATIO'] = data['GLOBAL_SUM_ANNUITY'] / data['AMT_INCOME_TOTAL']
    data['GLOBAL_REMAIN_FOR_LIFE'] = data['AMT_INCOME_TOTAL'] - data['GLOBAL_SUM_ANNUITY']
    data['GLOBAL_REMAIN_FOR_LIFE_PER_PERSON'] = data['GLOBAL_REMAIN_FOR_LIFE'] /data['CNT_FAM_MEMBERS']
    data['GLOBAL_CNT_ACTIVE_LOAN'] = data['CNT_ACTIVE_LOAN'] + data['BUR_CNT_ACTIVE_LOAN']
    return data
```

```python
def data_onehot(data):

    dummies = pd.get_dummies(data.NAME_EDUCATION_TYPE, prefix='EDUCATION',drop_first=True)   
    data = pd.concat([data, dummies], axis=1)
    
    dummies = pd.get_dummies(data.FLAG_OWN_REALTY, drop_first=True, prefix='REALTY')
    data = pd.concat([data, dummies], axis=1)
     
    dummies = pd.get_dummies(data.FLAG_OWN_CAR, drop_first=True, prefix='CAR')
    data = pd.concat([data, dummies], axis=1)
    
    dummies = pd.get_dummies(data.NAME_INCOME_TYPE, prefix='INCOME')
    dummies['INCOME_Pensioner_other'] = dummies.INCOME_Student + dummies['INCOME_Maternity leave'] + dummies['INCOME_Unemployed']
    del dummies['INCOME_Maternity leave']
    del dummies['INCOME_Student']
    del dummies['INCOME_Unemployed']
    data = pd.concat([data, dummies], axis=1)
    
    dummies = pd.get_dummies(data.NAME_FAMILY_STATUS, prefix='FAMILY')
    data = pd.concat([data, dummies], axis=1)
    
    dummies = pd.get_dummies(data.NAME_HOUSING_TYPE, prefix='HOUSING')
    data = pd.concat([data, dummies], axis=1)
    
    dummies = pd.get_dummies(data.CODE_GENDER, prefix='GENDER',drop_first=True)
    data = pd.concat([data, dummies], axis=1)
    
    data.columns = data.columns.str.replace(" ", "_")
    data.columns = data.columns.str.replace("/", "_")
    return data
```

```python
def traitement_global():
    print("traitement app_train")
    data = traitement_app_train()
    print("traitement bureau")
    data= traitement_bureau(data)
    print("traitement previous_application")
    data = traitement_previous_application(data)
    print("traitement installments_payments")
    data = traitement_installments_payments(data)
    print("traitement POS_CASH")
    data = traitement_POS_CASH_balance(data)
    print('traitement credit_card_balance')
    data = traitement_credit_card_balance(data)
    data = agg_features(data)
    data = data_onehot(data)
    return data
```

```python
def imputation_mediane(data):
    #imputation par la mediane des variables EXT_SOURCE
    imputer1 = SimpleImputer(missing_values=np.nan, strategy='median')
    data_median_imputer=imputer1.fit_transform(data[['EXT_SOURCE_1','EXT_SOURCE_2','EXT_SOURCE_3']])
    data_median_imputer = pd.DataFrame(data_median_imputer,columns = ['EXT_SOURCE_1','EXT_SOURCE_2','EXT_SOURCE_3'] )
    data[['EXT_SOURCE_1','EXT_SOURCE_2','EXT_SOURCE_3']] = data_median_imputer
    data['EXT_SOURCE_SUM'] = data['EXT_SOURCE_1']+data['EXT_SOURCE_2']+data['EXT_SOURCE_2']
    data['EXT_SOURCE_MEAN'] = (data['EXT_SOURCE_1']+data['EXT_SOURCE_2']+data['EXT_SOURCE_2'])/3
    return data
```
